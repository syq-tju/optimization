```python
def find_global_minimum(arr):
    if not arr:  # 如果数组为空
        return None, None
    min_value = arr[0]
    min_index = 0
    for i in range(1, len(arr)):
        if arr[i] < min_value:
            min_value = arr[i]
            min_index = i
    return min_index, min_value

# 示例数据
arr = [8, 4, 6, 3, 9, 5, 1, 2, 4, 0]
min_index, min_value = find_global_minimum(arr)
print(f"Global minimum at index {min_index} with value {min_value}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 6*x**2 + 9*x + 1

def golden_section_search(f, a, b, tol=1e-5):
    phi = (1 + 5**0.5) / 2  # 黄金比
    c = b - (b - a) / phi
    d = a + (b - a) / phi
    while abs(b - a) > tol:  # 当区间宽度大于容忍度时继续
        if f(c) < f(d):
            b = d
        else:
            a = c
        # 重新计算c和d
        c = b - (b - a) / phi
        d = a + (b - a) / phi

    return (a + b) / 2

# 寻找极小值
result = golden_section_search(f, 0, 4)

# 绘图
x = np.linspace(0, 4, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="f(x) = x^3 - 6x^2 + 9x + 1")
plt.scatter([result], [f(result)], color='red', zorder=5)
plt.text(result, f(result), f"Min at x={result:.3f}, f(x)={f(result):.3f}", fontsize=12, ha='left')
plt.title("Plot of the function with the minimum marked")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
